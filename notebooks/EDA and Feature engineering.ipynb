## Attribute information

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
dataset.head()
```

```python
dataset.info()
```

## Data cleaning

```python
#missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
dataset
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
```

```python
dataset
```

```python
dataset = dataset.dropna().reset_index(drop=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
#we need to drop inorder to combine
dataset=dataset.drop(122).reset_index(drop=True)
dataset
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
#remove unnecessary spaces
dataset.columns=dataset.columns.str.strip()
dataset.columns
```

## Change the required columns as int

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
dataset.info()
```

## Convert other columns into float(general way)

```python
objects=[col for col in dataset.columns if dataset[col].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        dataset[i]=dataset[i].astype(float)
```

```python
dataset.info()
```

```python
objects
```

```python
dataset.describe()
```

```python
dataset
```

```python
#save this cleaned dataset
dataset.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
df=dataset.drop(['day','month','year'],axis=1) #they are not required
df
```

```python
df['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df
```

```python
df['Classes'].value_counts()
```

```python
#plot density plot for all features
plt.style.use('seaborn-v0_8')
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# percentage for pie chart
percentage=df['Classes'].value_counts(normalize=True)*100
percentage
```

```python
#plotting pie chart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of classes')
```
